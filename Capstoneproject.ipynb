```python
#Scatter plot crime hotspots based on locations
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
plt.scatter(df['LON'], df['LAT'], alpha=0.5, marker='.', color='purple')
plt.title("Crime hotspots")
plt.xlabel('LONGITUDE') 
plt.ylabel('LATITUDE')
plt.grid(True)
plt.show()
```

```python
#Age wise crime distribution
plt.figure(figsize=(10,8))
sns.histplot(df['Vict_Age'],bins=25,)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title("Age wise crime distribution")
plt.show()
```

```python
#Gender wise Crime distribution
from matplotlib import pyplot as plt
import numpy as np
plt.figure(figsize=(10,5))
sns.histplot(df['Vict_Sex'],bins=20,)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title("Gender wise crime distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the Status dictionary
Status = {'AO': 27, 'IC': 447, 'AA': 24}

# Extract sector labels and counts
sector_labels = list(Status.keys())
sector_values = list(Status.values())

# Create the pie chart
plt.pie(sector_values, labels=sector_labels, autopct="%1.1f%%")
plt.title("Pie Chart for Status Report of crimes")
plt.show()

#bar chart 
plt.figure(figsize=(10,5))
sns.histplot(df['Vict_Sex'],bins=20,)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title("Gender wise crime distribution")
plt.show()
```
